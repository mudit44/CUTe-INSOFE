```python
def checkprime(n):
    flag = 0
    if(n == 2):
        print("Prime")
    elif(n%2 == 0):
        print("Not Prime")
    else:
        for i in range(3,(n-1)):
            if(n % i == 0):
                flag = 1
            
        if(flag == 1):
            print("Not Prime")
        else:
            print("Prime")
                
```

```python
checkprime(31)
```
